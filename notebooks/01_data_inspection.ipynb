```python
import os
import cv2

tb_dir = "../data/raw/TB"
normal_dir = "../data/raw/Normal"

tb_img = cv2.imread(os.path.join(tb_dir, os.listdir(tb_dir)[0]), cv2.IMREAD_GRAYSCALE)
normal_img = cv2.imread(os.path.join(normal_dir, os.listdir(normal_dir)[0]), cv2.IMREAD_GRAYSCALE)

print("TB image shape:", tb_img.shape)
print("Normal image shape:", normal_img.shape)
```

```python
sizes = set()

for folder in [tb_dir, normal_dir]:
    for img_name in os.listdir(folder)[:20]:
        img = cv2.imread(os.path.join(folder, img_name), cv2.IMREAD_GRAYSCALE)
        sizes.add(img.shape)

sizes
```
