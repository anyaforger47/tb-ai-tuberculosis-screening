```python
import os
import shutil
from sklearn.model_selection import train_test_split
```

```python
processed_tb_dir = "../data/processed/TB"
processed_normal_dir = "../data/processed/Normal"
base_split_dir = "../data/split"
```

```python
for split in ["train", "val", "test"]:
    for cls in ["TB", "Normal"]:
        os.makedirs(os.path.join(base_split_dir, split, cls), exist_ok=True)
```

```python
tb_images = os.listdir(processed_tb_dir)
normal_images = os.listdir(processed_normal_dir)

print("TB:", len(tb_images))
print("Normal:", len(normal_images))
```

```python
tb_train, tb_temp = train_test_split(
    tb_images, test_size=0.3, random_state=42
)
tb_val, tb_test = train_test_split(
    tb_temp, test_size=0.5, random_state=42
)
```

```python
normal_train, normal_temp = train_test_split(
    normal_images, test_size=0.3, random_state=42
)
normal_val, normal_test = train_test_split(
    normal_temp, test_size=0.5, random_state=42
)
```

```python
def copy_files(file_list, src_dir, dst_dir):
    for file in file_list:
        shutil.copy(
            os.path.join(src_dir, file),
            os.path.join(dst_dir, file)
        )
```

```python
copy_files(tb_train, processed_tb_dir, "../data/split/train/TB")
copy_files(tb_val, processed_tb_dir, "../data/split/val/TB")
copy_files(tb_test, processed_tb_dir, "../data/split/test/TB")
```

```python
copy_files(normal_train, processed_normal_dir, "../data/split/train/Normal")
copy_files(normal_val, processed_normal_dir, "../data/split/val/Normal")
copy_files(normal_test, processed_normal_dir, "../data/split/test/Normal")
```

```python
for split in ["train", "val", "test"]:
    tb_count = len(os.listdir(f"../data/split/{split}/TB"))
    normal_count = len(os.listdir(f"../data/split/{split}/Normal"))
    print(split, "TB:", tb_count, "Normal:", normal_count)
```
